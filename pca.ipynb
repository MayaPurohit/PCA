**Question 7:** How does the the PCA projected version of the mystery dataset compare to the projection that you achieved onto the `x`-`y` plane after rotating the dataset? **Why is there this similarity or difference in the results?**

**Answer 6:** 

We should drop the third principle component (the PC with the highest number and the lowest variance) because it does not contribute a significant amount of variance to the plot. Without the third principle component, we have 72% of the variance accounted for by the first two principle components. The third PC is not as influential in the dataset, meaning that it can be dropped. 

**Answer 7:** 

The two results are very similar to each other. This makes sense, because we used a similar process of min-max normalization, centering, rotation, and projection to get the final spiral image in the plane. There is a slight difference in the rotation that the final images have undergone. This may be due to the fact that we don't find the eigenvalues and eigenvectors when doing the data transformation without PCA. Finding the eignevectors will help us precisely find the major and minor axes for the data. The data transformations without PCA assume that the major and minor axes are on the x and y-axis and a rotation has to be done manually to find the best rotation angle that will capture the most variance. However, PCA will be more precise in determining the major and minor axes for the data and will capture as much of the variance as possible (using the covariance matrix). The PCs it finds will be more precise than the rotation we manually perform on the data. When doing PCA, we do not have to determine the angle that we will rotate by because the eigenvectors determine the best angle to rotate by for us. PCA fill find the optimal angle and return the result that captures the most variance.
